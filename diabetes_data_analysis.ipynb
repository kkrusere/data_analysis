```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/kkrusere/data_analysis/main/Data/diabetes.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
import plotly.express as px
```

```python
fig = px.histogram(df, x = "Pregnancies", color="Outcome")
fig.show(renderer='png', height=700, width=1000)
```

```python
fig = px.histogram(df, x = "Glucose", color="Outcome")
fig.show(renderer='png', height=700, width=1000)
```

```python
fig = px.scatter(df, y = "DiabetesPedigreeFunction", x = "Insulin", color = "Outcome" )
```

```python
no_zero_df = df[df['Insulin'] != 0]
no_zero_df.reset_index(drop=True, inplace= True)
fig = px.scatter(no_zero_df, y = "DiabetesPedigreeFunction", x = "Insulin", color = "Outcome" )
fig.show(renderer='png', height=700, width=1000)
```

```python
fig = px.histogram(no_zero_df, x = "Age", color="Outcome")
fig.show(renderer='png', height=700, width=1000)
```

```python
fig = px.histogram(no_zero_df, x = "Age")
fig.show(renderer='png', height=700, width=1000)
```

```python
import seaborn as sns
```
